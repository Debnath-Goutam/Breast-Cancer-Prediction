# Reading and Understanding Data

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('breast-cancer-wisconsin-data.csv')
data.head(20)
```

```python
data.info()
```

# Data Cleaning

Removing the **id** column since it is unnecessary for analysis and prediction

```python
data.drop(columns=['id'], inplace=True)
```

From the above summary of the dataset, we can observe that there are no missing data. Neither are there any features that have incorrect format of data. Hence there is no requirement for cleaning of data in this case.

```python
# Seperating the independent variables(X) and dependent variable(y)
y = data.pop('diagnosis')
X = data
```

```python
plt.figure(figsize=(20, 35))
for i, feature in enumerate(X):
    plt.subplot(8, 4, i+1)
    sns.histplot(X[feature], bins=30, kde=True)
plt.show()
```

```python
plt.figure(figsize=(20, 35))
for i, feature in enumerate(X):
    plt.subplot(8, 4, i+1)
    sns.distplot(X[feature])
plt.show()
```

From the above collection of plots, we can see the following features to show normal or almost normal distribution:
- **radius_mean**
- **texture_mean**
- **perimeter_mean**
- **area_mean**
- **smoothness_worst**
- **texture_worst**
Remaining all the features show a right skewed distribition with **concavity_se**, **concave points_se**, **perimeter_se, area_se** and **fractal_dimension_se** showing the highest positive skewness.

## Bivariate Analysis

```python
plt.figure(figsize=(20, 35))
for i, feature in enumerate(X):
    plt.subplot(8, 4, i+1)
    sns.boxplot(x=y, y=X[feature])
plt.show()
```

From the above collection of graphs, we can clearly see the most of the features have a significantly higher value for malignant. **radius_worst**, **radius_mean**, **perimeter_mean**, **area_mean**, **concavity_mean** and **permeter_worst** are some of the features that have very high values for malignant diagnosis than benign diagnosis.

**symmetry-se**, **smoothness_se**, **texture_se**, and **fractal_dimension_mean** appear to be the only features that have equal or less median value for malignant diagnosed cases than benign diagnosed cases

## Multivariate Analysis
